# FRUIT IMAGE PROCESSING

```python
#importing matplotlib libraray
import matplotlib.pyplot as plt
```

```python
#uploading image and loading it
x = plt.imread('fruit.jpg')

#viewing of image
plt.imshow(x)
```

```python
#dimension or shape of image x
x.shape
```

```python
#image is a combination of layer red(R) ,green(G), blue(B) stack over each other
#as shown in sahpe (408,612,3).. here 3 justify it
R_img,B_img,G_img = x.copy(),x.copy(),x.copy()
```

####  1. IMAGE IN ONLY RED COLOUR

```python
#lets view image only in RED by making other axis zero
R_img[:,:,[1,2]] = 0
plt.imshow(R_img)
```

#### 2. IMAGE IN GREEN COLOUR

```python
#lets view image only in GREEN by making other axis zero
G_img[:,:,[0,2]] = 0
plt.imshow(G_img)
```

#### 3. IMAGE IN ONLY BLUE COLOUR

```python
#lets view image only in BLUE by making other axis zero
B_img[:,:,[0,1]] = 0
plt.imshow(B_img)
```
